1-Lists and Arrays

Define and manipulate lists.

A list is a data structure that allows you to store an ordered collection of items. These items can be of different data types (integers, strings, etc.). Lists are mutable, meaning you can change their content after they are created.

In Python, lists are defined by enclosing the items in square brackets ([]), with each item separated by a comma.

```python
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)
```

```python
my_list = [1, 2, 3, "apple", 4.5]
```

**Operations on Lists:**

```python
import array

# List operations
my_list = [1, 2, 3, "apple", 4.5]
print("Original list:", my_list)

# Accessing an element in the list
print("First element of list:", my_list[0])

# Adding elements to the list
my_list.append("banana")  # Adds 'banana' to the end of the list
print("After append:", my_list)

my_list.insert(1, "orange")  # Inserts 'orange' at index 1
print("After insert:", my_list)

# Removing elements from the list
my_list.remove("apple")  # Removes 'apple' from the list
print("After remove:", my_list)

# Popping an element from the list (removes and returns it)
popped_item = my_list.pop(0)  # Removes and returns the item at index 0
print("Popped item:", popped_item)
print("After pop:", my_list)

# List slicing (sublist)
sub_list = my_list[1:3]  # Extracts elements from index 1 to 2
print("Sublist (sliced):", sub_list)

# Length of the list
print("Length of the list:", len(my_list))


# Array operations (requires array module)
my_array = array.array('i', [1, 2, 3, 4])
print("\nOriginal array:", my_array)

# Accessing an element in the array
print("First element of array:", my_array[0])

# Adding elements to the array
my_array.append(5)  # Adds 5 to the end of the array
print("After append:", my_array)

my_array.insert(1, 6)  # Inserts 6 at index 1
print("After insert:", my_array)

# Removing elements from the array
my_array.remove(3)  # Removes the first occurrence of 3
print("After remove:", my_array)

# Array slicing (subarray)
sub_array = my_array[1:3]  # Extracts elements from index 1 to 2
print("Subarray (sliced):", sub_array)

# Length of the array
print("Length of the array:", len(my_array))
```

**Tuples and Dictionaries**

A tuple is an ordered, immutable collection of elements. Unlike lists, tuples cannot be modified after they are created. They can contain elements of different data types and are defined using parentheses ()

Key characteristics:


*   Immutable: Once a tuple is created, you cannot change, add, or remove elements.
*   Ordered: Elements have a defined order, and this order will not change.

*   Immutable: Once a tuple is created, you cannot change, add, or remove elements.
*   Indexed: You can access elements by their index (starting from 0).
Can store heterogeneous data (data of different types)






```python
## Define a tuple
person = ("John", 25, "Engineer")

# Access elements by index
print(person[0])  # Output: John
print(person[1])  # Output: 25

# Tuples are immutable, so you cannot modify an element
# person[1] = 30  # This will raise an error

# Concatenate two tuples
new_person = person + ("New York",)
print(new_person)  # Output: ('John', 25, 'Engineer', 'New York')
```

**Dictionaries:**


A dictionary is an unordered collection of data stored in key-value pairs. Each key is unique, and values can be of any data type. Dictionaries are mutable, meaning you can add, modify, or remove items.

Key characteristics:


*   Unordered: Items are not stored in a specific order.

*   Key-Value Pair: Each item is represented as a key-value pair.
*   Mutable: You can add, modify, or remove key-value pairs.


*   Keys are unique: Each key must be unique, but values can be repeated.


```python
# Define a dictionary
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

# Access values using keys
print(student["name"])  # Output: Alice
print(student["age"])   # Output: 20

# Add a new key-value pair
student["city"] = "Paris"

# Modify an existing value
student["grade"] = "B"

# Remove a key-value pair
del student["age"]

# Check if a key exists
print("city" in student)  # Output: True

# Print all keys and values
print(student.keys())  # Output: dict_keys(['name', 'grade', 'city'])
print(student.values())  # Output: dict_values(['Alice', 'B', 'Paris'])
```

```python
# Define the functions for each operation
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."

# Mapping operations to functions using a dictionary
operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide
}

# Get user input
print("Select operation:")
print(" + for addition")
print(" - for subtraction")
print(" * for multiplication")
print(" / for division")

operation = input("Enter operation: ")

# Get the numbers from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform the operation using the dictionary and functions
result = operations.get(operation, lambda x, y: "Invalid operation")(num1, num2)

# Display the result
print(f"The result is: {result}")
```
