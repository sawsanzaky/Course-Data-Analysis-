**Matplotlib**

To create a plot using the matplotlib library in Python, you can start by setting up a plotting session and using various functions like plot(), bar(), scatter(), etc.

**Example of a line plot using plot():**

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create the plot
plt.plot(x, y)

# Add title to the plot
plt.title('Simple Plot')

# Add labels to the axes
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

**Explanation of the code:**
plt.plot(x, y): Plots a line passing through the points defined by x and y.

*   plt.title(): Adds a title to the plot.
*   plt.xlabel() and plt.ylabel(): Adds labels to the X and Y axes.


*   plt.show(): Displays the plot.





**Another example using bar() to create a bar chart:**

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]

# Create a bar chart
plt.bar(categories, values)

# Add title
plt.title('Bar Chart')

# Add labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

**Seaborn**

Seaborn is a statistical data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn simplifies many of the complexities of Matplotlib and is often preferred for its ability to work directly with pandas DataFrames.

**1. Line Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a Seaborn line plot
sns.lineplot(x=x, y=y)

# Add title and labels
plt.title('Line Plot using Seaborn')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

**2. Bar Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 30]

# Create a Seaborn bar plot
sns.barplot(x=categories, y=values)

# Add title and labels
plt.title('Bar Plot using Seaborn')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
```

**3. Scatter Plot:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Create a Seaborn scatter plot
sns.scatterplot(x=x, y=y)

# Add title and labels
plt.title('Scatter Plot using Seaborn')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

**4. Heatmap (often used for correlation matrices):**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Example correlation matrix
data = np.random.rand(10, 12)
corr_matrix = np.corrcoef(data)

# Create a Seaborn heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title('Heatmap using Seaborn')

# Show the plot
plt.show()
```

**5. Histogram:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create a Seaborn histogram
sns.histplot(data, kde=True)

# Add title
plt.title('Histogram using Seaborn')

# Show the plot
plt.show()
```
