```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'Score': [85, 90, 78, 92, 88]
}
```

```python
# تحويل القاموس إلى DataFrame
df = pd.DataFrame(data)
```

```python
# رسم بياني باستخدام Matplotlib
plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Score'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Score by Name')
plt.show()
```

```python
# رسم بياني باستخدام Seaborn
sns.barplot(x='Name', y='Score', data=df, palette='coolwarm')
plt.title('Seaborn: Score by Name')
plt.show()
```

```python
# رسم بياني باستخدام Plotly
fig = px.bar(df, x='Name', y='Score', title='Plotly: Score by Name', color='Score')
fig.show()
```
